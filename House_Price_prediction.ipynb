```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
```

```python
path="C:/Users/VenkatBhaskar/Downloads/train.csv"
path1="C:/Users/VenkatBhaskar/Downloads/test.csv"
df=pd.read_csv(path)
fd=pd.read_csv(path1)
```

```python
df
```

```python
frames=[df,fd]
res=pd.concat(frames)
res
```

```python
pd.isnull(res).sum()[pd.isnull(res).sum()>0]
```

```python
#Removing attributes having more than 50% null values
res=res.drop(columns=['Id','Alley','PoolQC','FireplaceQu','Fence','MiscFeature'])
res
```

```python
res.isnull().sum()[res.isnull().sum()>0]
```

```python
y=df['SalePrice']
res=res.drop(columns=['SalePrice'])
res
```

```python
#Filling Missing values
#Interpolate function uses existing values in the dataframe to estimate the missing rows
#res1=res.interpolate(method='linear',limit_direction='forward')
#res1
```

```python
#res1.isnull().sum()[res1.isnull().sum()>0]
```

```python
c1=res.select_dtypes(include=['object']).columns.tolist()
c2=res.select_dtypes(exclude=['object']).columns.tolist()
```

```python
#Replacing missing values with most occuring value
for i in c1:
    res[i].fillna(res[i].mode()[0],inplace=True)
```

```python
res
```

```python
res.isnull().sum()[res.isnull().sum()>0]
```

```python
for i in c2:
    res[i].fillna(res[i].mean(),inplace=True)
```

```python
res
```

```python
res.isnull().sum()[res.isnull().sum()>0]
```

```python
a=pd.get_dummies(res,columns=c1)
a
```

```python
#Splitting features into test and train
X_train=a.iloc[:1460,:]
X_test=a.iloc[1460:,:]
```

```python
y
```

```python
model=XGBClassifier()
model.fit(X_train,y)
```

```python
y_pred=model.predict(X_test)
y_pred
```
